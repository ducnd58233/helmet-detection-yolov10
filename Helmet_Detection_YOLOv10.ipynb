#Fine-Tuning

```python
%cd /content/
!gdown '1twdtZEfcw4ghSZIiPDypJurZnNXzMO7R'
!mkdir safety_helmet_dataset
!unzip -q '/content/Safety_Helmet_Dataset.zip' -d '/content/safety_helmet_dataset'
```

```python
import torch

YAML_PATH = '/content/safety_helmet_dataset/data.yaml'
EPOCHS = 10
IMG_SIZE = 640
BATCH_SIZE = 16

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)

model.train(data=YAML_PATH, epochs=EPOCHS, imgsz=IMG_SIZE, batch=BATCH_SIZE)
```

##Evaluate Fine-tuning model

```python
TRAINED_MODEL_PATH='/content/yolov10/runs/detect/train/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)
model.val(data=YAML_PATH,imgsz=IMG_SIZE,split='test')
```

```python
from pytube import YouTube

YOUTUBE_VIDEO_PATH = 'https://youtu.be/wqPSsu7XQ74'
yt = YouTube(YOUTUBE_VIDEO_PATH)
stream = yt.streams.filter(file_extension='mp4').first()
video_path = stream.download('/content')

# Make predictions on the video
results = model(source=video_path, stream=True, save=True, show=True)

# Extract the outputs
for result in results:
    outputs = result.boxes.xyxy
    print(outputs)
```
